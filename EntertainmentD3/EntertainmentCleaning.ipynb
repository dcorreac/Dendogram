```python
import pandas
from IPython.display import display, HTML


df = pandas.read_csv('Entertainment.csv')

display(df)
```

We are only interested in the "Name", "Profession", and "Company" columns, so let's remove the rest. 


```python
df = df.drop('Food and beverage category', 1)
df = df.drop('Brands', 1)
df = df.drop('Ethnicity', 1)
df = df.drop('Age', 1)
df = df.drop('Links', 1)
df = df.drop('Notes', 1)






display(df)
```

Now that we have narrowed down the data to obtaining the values we want,we must structure this in a way such that hierarchy is apparent and easily accessible when we create the hierarchal JSON file. I will do this by two different dictionaries, one in which Profession is a key in which we will have companies as the associated values. 
The second dictionary will have Company as a key with respective associated names. 


```python

        

master = dict()         
for index, rows in df.iterrows(): 
    profession = rows['Profession']
    Company_Names = dict()
    for indes, rows2 in df.iterrows():
        profession2 = rows2['Profession']
        company = rows2['Company']
        name = rows2['Name']
        if profession2 == profession:
            if company in Company_Names: 
                if name not in Company_Names[company]:
                    Company_Names[company].append(name)
            else: 
                Company_Names[company] = [name]
    master[profession] = Company_Names

print(master['Actress']['Borden Milk'])
```

Now that we have both of the dictionaries, let's make this into a JSON file wi 

```python
import json 

js =json.dumps(master)

# Open new json file if not exist it will create
fp = open('Winb.json', 'a')
parsed = json.loads(js)
js = json.dumps(parsed, indent=4, sort_keys=True)
# write to json file
fp.write(js)

# close the connection
fp.close()
```
